# Program 1

```python
#Creating the driver
driver=webdriver.Chrome(r"C:\Users\maahi\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#opening the desired page
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
#Creating the rank field
rank1=[] 
# fetching  of details for rank
try:
    rank_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in rank_tag[0:30]:
        rank1.append(i.text)
except NoSuchElementException:
    rank1.append('-')
```

```python
#Creating the upload field
upload=[] 
# fetching  of details for upload
try:
    up_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in up_tag[0:30]:
        upload.append(i.text)
except NoSuchElementException:
    upload.append('-')
```

```python
#Creating the view field
views=[] 
# fetching  of details for views
try:
    view_tag=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in view_tag[0:30]:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')
```

# Program10
Details of Datasets from UCI machine learning repositories. Ur = https://archive.ics.uci.edu/
You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year

```python
#Accessing the driver
driver=webdriver.Chrome(r"C:\Users\maahi\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#Opening the desired page
driver.get("https://archive.ics.uci.edu/")
```

```python
#Clicking on view-all-datasets
view_tab=driver.find_element(By.XPATH,'//a[@href="datasets.php"]').click()
```

```python
#Creating the attribute list
dataset=[] 
dtype=[] 
task=[] 
attr_type=[] 
instances=[] 
attributes=[] 
year=[]

# Scrapping info for Dataset Name
try:
    datas_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in datas_tag:
        dataset.append(i.text)
except NoSuchElementException:
    dataset.append('-')

# Scrapping info for Data Type
try:
    dtype_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
    for i in dtype_tag:
        dtype.append(i.text)
except NoSuchElementException:
    dtype.append('-')

# Scrapping info for task
try:
    task_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
    for i in task_tag:
        task.append(i.text)
except NoSuchElementException:
    task.append('-')

# Scrapping info for attribute_type
try:
    attr_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
    for i in attr_tag:
        attr_type.append(i.text)
except NoSuchElementException:
    attr_type.append('-')

# Scrapping info for instances
try:
    ins_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
    for i in ins_tag:
        instances.append(i.text)
except NoSuchElementException:
    instances.append('-')

# Scrapping info for no. of attributes
try:
    nattr_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
    for i in nattr_tag:
        attributes.append(i.text)
except NoSuchElementException:
    attributes.append('-')

# Scrapping info for year
try:
    year_tag=driver.find_elements(By.XPATH,"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
    for i in year_tag:
        year.append(i.text)
except NoSuchElementException:
    year.append('-')
```

```python
print(len(dataset),len(dtype),len(task),len(attr_type),len(instances),len(attributes),len(year))
```

```python
#Creating DataFrame
df10=pd.DataFrame({'Dataset Name':dataset,'Data Type':dtype[1:624],'Task':task[1:624],'Attribute Type':attr_type[1:624],'No of Instances':instances[1:624],'No of attribute':attributes[1:624],'Year':year[1:624]})
print("Details of Datasets from UCI Machine Learning Repositories")
df10
```
