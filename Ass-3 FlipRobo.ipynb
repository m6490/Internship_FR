```python
#Creating Attribute List
hostel_name=[]
distance=[]
comb_prices=[]
pvt_prices=[]
dorms_price=[]
rating=[]
tot_reviews=[]
over_all_re=[]
facilities=[]
description=[]
url=[]
```

```python
#Since there are total 3 pages on the website
for j in range(0,3):
    #hostel name data
    try:
        name_tag=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
        for i in name_tag:
            hostel_name.append(i.text)
    except NoSuchElementException:
           hostel_name.append('-')
            
    #distance data
    try:
        dis_tag=driver.find_elements(By.XPATH,'//span[@class="description"]')
        for i in dis_tag:
            distance.append(i.text)
    except NoSuchElementException:
        distance.append('-')
        
    #pvt_prices data
    
    try:
        pvt_tag=driver.find_elements(By.XPATH,'//a[@class="prices"]//div[1]//div')
        for i in pvt_tag: 
            pvt_prices.append(i.text)
    except NoSuchElementException:
        pvt_prices.append('-')       
        
    #dorm_prices data
    try:
        dorm_tag=driver.find_elements(By.XPATH,'//a[@class="prices"]//div[2]//div')
        for i in dorm_tag:
            dorms_price.append(i.text)
    except NoSuchElementException:
        dorms_price.append('-')
        
    #rating data
    try:
        ra_tag=driver.find_elements(By.XPATH,'//div[@class="score orange big" or @class="score gray big"]')
        for i in ra_tag:
            rating.append(i.text)
    except NoSuchElementException:
        rating.append('-')
    
    #reviews data
    try:
        re_tag=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
        for i in re_tag:
            tot_reviews.append(i.text)
    except NoSuchElementException:
        tot_reviews.append('-')
        
    #overall review data
    try:
        o_tag=driver.find_elements(By.XPATH,'//div[@class="keyword"]//span')
        for i in re_tag:
            over_all_re.append(i.text)
    except NoSuchElementException:
        over_all_re.append('-')
        
    #facilities data
    try:
        fa_tag=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
        for i in fa_tag:
            facilities.append(i.text)
    except NoSuchElementException:
        facilities.append('-')
        
    #description data
    try:
        de_tag=driver.find_elements(By.XPATH,'//div[@class="rating-factors prop-card-tablet rating-factors small"]')
        for i in de_tag:
            description.append(i.text)
    except NoSuchElementException:
        description.append('-')
    
    #url data
    try:
        url_tag=driver.find_elements(By.XPATH,'//div[@class="prices-col"]//a[2]')
        for i in url_tag:
            url.append(i.get_attribute('href'))
    except NoSuchElementException:
        url.append('-')
    
    #next_button=driver.find_elements(By.XPATH,'//div[@class="pagination-item pagination-next""]')
   # next_button.click()
    #time.sleep(4)
print("Information Scrapped")
```
